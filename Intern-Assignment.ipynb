# b) Draw time series plots with anomaly regions

```python
import pandas as pd
import matplotlib.pyplot as plt
def plot_with_anomaly(data, anomalies, title=''):
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data.values, label='Time Series Data')
    if 'start_date' in anomalies.columns and 'end_date' in anomalies.columns:
        for anomaly in anomalies.itertuples():
            plt.axvspan(anomaly.start_date, anomaly.end_date, color='red', alpha=0.3, label='Anomaly')
    
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_with_anomaly(test_data_1,label_data_1,'Time Series Data 1 with Anomaly Regions')
plot_with_anomaly(test_data_1,label_data_1,'Time Series Data 2 with Anomaly Regions')
plot_with_anomaly(test_data_1,label_data_1,'Time Series Data 3 with Anomaly Regions')
plot_with_anomaly(test_data_4,label_data_4,'Time Series Data 4 with Anomaly Regions')
```

# c) Perform EDA and find out root cause

```python
def perform_eda(test_data):
    summary_stats = test_data.describe()
    correlation_matrix = test_data.corr()
    return summary_stats, correlation_matrix
```

# d) Find out the variables which are the root cause for the anomaly

```python
def find_root_cause(test_data, label_data, threshold=0.5):
    test_data_sorted = test_data.sort_index()
    label_data_sorted = label_data.sort_values(by='0')
    merged_data = pd.merge_asof(test_data_sorted, label_data_sorted, left_index=True, right_on='0', direction='nearest')
    correlation_matrix = merged_data.corr()
    root_cause_variables = correlation_matrix[correlation_matrix['0'].abs() > threshold].index.tolist()
    
    return root_cause_variables
```

```python
root_cause_variables_1 = find_root_cause(test_data_1, label_data_1)
```

```python
root_cause_variables_2 = find_root_cause(test_data_2, label_data_2)
```

```python
def find_root_cause(test_data, label_data, threshold=0.5):
    test_data_sorted = test_data.sort_index()
    label_data_sorted = label_data.sort_values(by='timestamp_(min)')
    test_data_sorted.index = test_data_sorted.index.astype(float)
    merged_data = pd.merge_asof(test_data_sorted, label_data_sorted, left_index=True, right_on='timestamp_(min)', direction='nearest')
    correlation_matrix = merged_data.corr()
    root_cause_variables = correlation_matrix[correlation_matrix['label'].abs() > threshold].index.tolist()

    return root_cause_variables
```

```python
root_cause_variables_4 = find_root_cause(test_data_4, label_data_4)
```
